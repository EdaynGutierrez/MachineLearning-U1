```python
import numpy as np# Importamos la librería numpy y le asignamos el alias 'np' por convención
vector = np.array ([1,2,3,4,5]) # Creamos un vector (arreglo de 1 dimensión) con 5 elementos
matriz = np.array ([[1,2,3],[4,5,6]]) # Creamos una matriz (arreglo de 2 dimensiones) de 2 filas y 3 columnas
print(vector.ndim, vector.shape) # Imprimimos el número de dimensiones (ndim) y la forma (shape) del vector
print(matriz.ndim, matriz.shape)# Imprimimos el número de dimensiones y la forma de la matriz
```

```python
A = np.array([[1,2], [3,4]]) # Creamos dos matrices de 2x2
B= np.array([[5,6], [7,8]])

C=A+B# Sumamos las matrices A y B elemento por elemento
D= np.sqrt(A) # Calculamos la raíz cuadrada (sqrt) de cada elemento de la matriz A
# Imprimimos las matrices originales
print(A)
print(B)
```
