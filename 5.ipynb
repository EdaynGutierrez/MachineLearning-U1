```python
import numpy as np

# x: Vector de entrada con 3 características (datos que recibe la neurona)
x = np.array([1.0, 2.0, -1.0])

# w: Vector de pesos (weights) correspondientes a cada entrada
w = np.array([0.5, -0.3, 0.8])

# b: Valor de sesgo (bias) de la neurona
b = 0.1

# z: Cálculo lineal. Multiplica cada entrada por su peso (dot product) y suma el sesgo
z = np.dot(w, x) + b

# Imprimimos el resultado crudo antes de pasar por la función de activación
print("salida lineal: ", z)
```

```python
# [aso del resultado lineal a través de una función de activación (ReLU) para determinar si la señal de la neurona "pasa" o se apaga (se vuelve cero).
# Definimos la función ReLU que convierte cualquier número negativo en 0
def relu(z):
    return np.maximum(0, z)

# a: Salida "activada". Pasamos el resultado anterior (z) por la función ReLU
a = relu(z)

# Imprimimos el resultado final. Como 'z' era -0.8 (negativo), ReLU devuelve 0.0
print("salida activada: ", a)
```

```python
# Importamos Sequential: un contenedor para apilar capas de neuronas una tras otra
from tensorflow.keras.models import Sequential
# Importamos Dense: el tipo de capa estándar donde todas las neuronas se conectan entre sí
from tensorflow.keras.layers import Dense

# Creamos el modelo definiendo sus capas en una lista
model = Sequential([
    # Capa Oculta: 3 neuronas, usa activación 'relu'.
    # 'input_shape=(3,)' le dice a la red que espere recibir 3 datos de entrada.
    Dense(3, activation='relu', input_shape=(3,)),

    # Capa de Salida: 1 neurona (típico para predecir un resultado binario: sí/no).
    # Usa activación 'sigmoid', que comprime el resultado entre 0 y 1 (como una probabilidad).
    Dense(1, activation='sigmoid')
])

# Imprimimos un resumen de la arquitectura del modelo (capas, forma de salida y parámetros).
model.summary()
```
