```python
# Importamos los modelos desde la librería scikit-learn
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Imprimimos la clase base del modelo de Regresión Logística
print(LogisticRegression)

# Instanciamos una red neuronal (Perceptrón Multicapa) definiendo su estructura:
# hidden_layer_sizes=(64,54) crea dos capas ocultas con 64 y 54 neuronas respectivamente
print(MLPClassifier(hidden_layer_sizes=(64,54)))
```

```python
import numpy as np

# x: Vector de entrada (los datos que recibe la red)
x = np.array([1.0, 2.0])

# w: Matriz de pesos . Representa la importancia de cada conexión
w = np.array([[0.5, -1.0],
              [1.0, 0.5]])

# b: Vector de sesgo (bias). Desplaza el resultado para ajustar el modelo
b = np.array([0.1, -0.2])

# Calculamos la salida:
# np.dot() hace la multiplicación matricial (producto punto) entre pesos y entradas
# Luego sumamos el vector de sesgo (b)
y = np.dot(w, x) + b

# Imprimimos el resultado (la salida de esta neurona/capa)
print(y)
```

```python
# Definimos la función de activación ReLU
def relu(x):
    return np.maximum(0, x) #compara cada valor con 0 y devuelve el mayor

# Definimos un nuevo vector de entrada con un valor positivo y uno negativo
x = np.array([1.0, -1.0])

# Simulamos el avance de los datos a través de 3 capas consecutivas (bucle)
for i in range(3):
    # np.eye(2) crea una matriz identidad de 2x2, np.dot() multiplica esa matriz por la entrada 'x', relu() toma ese resultado y convierte el -1.0 en 0.0, El nuevo valor se guarda de nuevo en 'x' para la siguiente capa
    x = relu(np.dot(np.eye(2), x))

    # Imprimimos el resultado formateando el texto dinámicamente (f-string)
    print(f'capa {i+1}: {x}')
```
