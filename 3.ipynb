```python
# Importamos la librería pandas
import pandas as pd


# Creamos un diccionario de Python llamado 'datos'.
# Cada clave ("pixel_1", "pixel_2", "clase") se convertirá en el nombre de una columna,
# y las listas de números serán los valores de las filas.
datos= {
    "pixel_1":[120, 130, 125],
    "pixel_2":[60, 65, 70],
    "clase":[0, 1, 0]

}
# Convertimos el diccionario en un DataFrame
df= pd.DataFrame(datos)
# Convertimos el diccionario en un DataFrame y la forma shape
print(df)
print(df.shape)
```

```python

#filtrar solo clase 1
df_clase_1 = df[df["clase"] == 1]

#seleccionar columnas especificas
pixeles = df.loc[:, ["pixel_1", "pixel_2"]]

# Imprimimos la tabla filtrada y luego Imprimimos el subconjunto que solo contiene las columnas de píxeles (sin la columna "clase").
print(df_clase_1)
print(pixeles)
```

```python


## Normalizamos los valores de los píxeles dividiéndolos por 255.0.
pixeles_norm = pixeles / 255.0

#media por columna
media =pixeles.mean()

#imprimimos la tabla  y el valor promedio calculado para "pixel_1" y para "pixel_2"
print(pixeles_norm)
print(media)
```
