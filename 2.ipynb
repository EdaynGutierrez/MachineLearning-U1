```python
# Importamos numpy
import numpy as np
# Importamos el submódulo pyplot de matplotlib para crear las gráficas, con el alias 'plt'
import matplotlib.pyplot as plt

# ==========================================
# Gráfica 1: Subgráficos (Seno y Coseno)


# Generamos puntos espaciados entre 0 y 2*pi
x = np.linspace(0, 2*np.pi, 100)

# Creamos una figura (la ventana de la gráfica) con un tamaño específico de 8x4 pulgadas
plt.figure(figsize=(8,4))

# Creamos el primer subgráfico en una cuadrícula de 1 fila y 2 columnas (posición 1)
plt.subplot(1, 2, 1)
# Graficamos 'x' contra el seno de 'x'
plt.plot(x, np.sin(x))
# Le ponemos título a este subgráfico
plt.title("funcion seno")

# Creamos el segundo subgráfico en la misma cuadrícula de 1x2 (posición 2)
plt.subplot(1, 2, 2)
# Graficamos 'x' contra el coseno de 'x'
plt.plot(x, np.cos(x))
# Le ponemos título al segundo subgráfico
plt.title("funcion coseno")




# Gráfica 2: Función Cuadrática
# Redefinimos 'x': 100 puntos desde -10 hasta 10
x = np.linspace(-10, 10, 100)
# Definimos 'y' como el cuadrado de 'x' (x^2)
y = x**2

# Creamos una nueva figura
plt.figure()
# Graficamos la función. 'label' asigna una etiqueta que se usará en la leyenda
plt.plot(x, y, label="y = x") # Nota: la etiqueta dice "y=x" pero la función es cuadrática
# Añadimos un título principal y Nombramos el eje horizontal (X) y el eje vertical (Y)
plt.title("funcion cuadratica")
plt.xlabel("x")
plt.ylabel("y")
# Mostramos la leyenda (el recuadro que identifica qué es cada línea según su 'label')
plt.legend()
# Activamos la cuadrícula de fondo para facilitar la lectura
plt.grid(True)
# Mostramos todas las figuras que se han creado hasta este punto
plt.show()

# ==========================================
# Gráfica 3: Función Exponencial y Guardado


# Redefinimos 'x': 50 puntos desde 0 hasta 5
x = np.linspace(0, 5, 50)
# Calculamos la exponencial de 'x' (e^x)
y = np.exp(x)

# Creamos una nueva figura
plt.figure()
# Graficamos los datos
plt.plot(x, y)
# Añadimos título y nombres a los ejes
plt.title("crecimiento exponencial")
plt.xlabel("x")
plt.ylabel("e^x")

# Guardamos la gráfica generada como una imagen (.png) en la misma carpeta del script
# 'dpi=300' asegura que la imagen guardada tenga una alta resolución
plt.savefig("grafica_exponencial.png", dpi=300)

# Mostramos esta última gráfica en pantalla
plt.show()


# Generamos un arreglo de 100 números espaciados uniformemente entre -10 y 10
# Esto servirá como nuestros datos para el eje horizontal
x = np.linspace(-10, 10, 100)

# Calculamos el valor de 'y' elevando cada elemento de 'x' al cuadrado (x^2)
# Esto servirá como nuestros datos para el eje vertical
y = x**2

# Inicializamos una nueva figura (la ventana donde se dibujará la gráfica)
plt.figure()

# Trazamos la línea de la gráfica usando los datos 'x' y 'y'.
# 'label' define el texto que aparecerá en la leyenda para identificar esta línea.
plt.plot(x, y, label='y=x^2')

# Añadimos un título principal en la parte superior de la gráfica
plt.title("Funcion cuadratica")

# Etiquetamos el eje horizontal y vertica; para saber qué representa
plt.xlabel("x")

plt.ylabel("y")

# Activamos la cuadrícula de fondo para facilitar la lectura de los puntos exactos
plt.grid(True)

# Mostramos la leyenda (el pequeño recuadro que mostrará 'y=x^2' según el 'label' de plt.plot)
plt.legend()

plt.show()
```
